```python
#pip install ipeadatapy
import ipeadatapy as idpy
import pandas as pd
from datetime import datetime
```

```python
# Apresentação de todas as variáveis Macroeconômicas disponíveis no Ipea Data
idpy.themes(macro=1)
```

```python
# As características desejadas das séries temporais podem ser alteradas aqui
test = idpy.metadata(big_theme = 'Macroeconômico', country = "BRA", frequency = "Mensal")
```

```python
# Filtragem de séries que foram atualizadas pela última vez pelo menos em dezembro de 2020 (séries recentes)

filt_lu = [0]*len(test)

# Last Update
for i in range(len(test)):
    ti = datetime.strptime(test["LAST UPDATE"][i][0:10], "%Y-%m-%d")
    if ti >= datetime.strptime("2020-12-01", "%Y-%m-%d"):
        filt_lu[i] = True
    else:
        filt_lu[i] = False

test_lu = test.iloc[filt_lu]
```
